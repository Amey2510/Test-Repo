```python
from warnings import filterwarnings
filterwarnings('ignore')
from keras.datasets import mnist
```

```python
(xtrain,ytrain),(xtest,ytest) = mnist.load_data() #mnist has all handwritten numbers from 0 to 9, we just have to call xtrain and xtest
                                                  # data is already in a split format
```

```python
print(xtrain.shape,xtest.shape,ytrain.shape,ytest.shape) # 28,28 is length and width
```

```python
import matplotlib.pyplot as plt
plt.imshow(xtrain[0],cmap='gray')
```

```python
import seaborn as sns
plt.figure(figsize=(25,25))
sns.heatmap(xtrain[10],cmap='gray',annot=True,fmt='d')
```

## Data Preprocessing

```python
## Since images can't be understood by the computer, we are going to standardize them in 0 to 1 range using formula img/255
```

```python
print(xtrain[0])
```

```python
xtrain_scl = xtrain.astype('float32')/255
xtest_scl = xtest.astype('float32')/255
```

```python
ytrain[0]
```

```python
xtrain_scl[0]
```

```python
from keras.utils import to_categorical
```

```python
ytrain_scl = to_categorical(ytrain)  # We convert Y labels using to categorical as when ytrain is shown in the output, the number is
ytest_scl = to_categorical(ytest)    # displayed and hence the classes from 0 to 9, so we encode them using to_Categorical
# Which will then display an array of 10 classes for every single input and 1 being highest will represent the number of that input.
```

```python
ytrain_scl[0]
```

```python
xtrain_scl.shape # Here we will have to increase the dimension to add channel, to mention whether the colocode is black_gray or rgb
```

```python
import numpy as np
xtrain_scl = np.expand_dims(xtrain_scl,axis=-1)
xtest_scl = np.expand_dims(xtest_scl,axis=-1)
```

```python
xtrain_scl.shape # Dimension Updated
```

```python
ytrain_scl.shape
```

## CNN Model Building

```python
from keras.models import Sequential
from keras.layers import Input,Dense,Conv2D,MaxPool2D,BatchNormalization,Dropout,Flatten

# Model Intialization
model = Sequential()

# Input Layer
model.add(Input(shape=(28,28,1),))

# Add Convolutional Layer
model.add(Conv2D(filters=32,kernel_size=(2,2),activation='relu'))
# Add MaxPooling layer
model.add(MaxPool2D(pool_size=(2,2)))


# Add Convolutional Layer 2
model.add(Conv2D(filters=64,kernel_size=(2,2),activation='relu'))
# Add MaxPooling layer
model.add(MaxPool2D(pool_size=(2,2)))

# Add flatten layer
model.add(Flatten())

# Add fullyconnceted layer
model.add(Dense(units=32,activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(rate=0.2))
model.add(Dense(units=10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping
```

```python
es = EarlyStopping(monitor='val_loss',patience=10)
```

```python
nn = model.fit(xtrain_scl,ytrain_scl,validation_split=0.2,epochs=200,callbacks=[es])
```

## Learnings Curve

```python
plt.plot(nn.history['loss'],label='Training Loss')
plt.plot(nn.history['val_loss'],label='Validation Loss')
plt.xlabel('Number of Epochs')
plt.ylabel('Categorical_Entropy_Loss')
plt.title('Learning Curve')
plt.legend()
plt.show()
```

## Evaluation Metrics

```python
ypreds = model.predict(xtest_scl)
ypreds[0]
```

```python
ypreds_final = np.argmax(ypreds,axis=1)
ypreds_final[0]
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
```

```python
print(classification_report(ytest,ypreds_final))
```

```python
ConfusionMatrixDisplay.from_predictions(ytest,ypreds_final)
```
