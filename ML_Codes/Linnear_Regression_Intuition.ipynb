```python
import numpy as np
```

```python
bills = np.array([50,32,78,99,12,25,80,65])
tips = np.array([4,3,6,9,1,2,7,5])
```

```python
b_mean = bills.mean()
t_mean = tips.mean()
b_mean,t_mean
```

```python
slope = np.sum((bills-np.mean(bills))*(tips-np.mean(tips)))/np.sum(np.square(bills-np.mean(bills)))
slope
```

```python
intercept = (t_mean)-(b_mean)*slope
```

```python
intercept
```

```python
pred_tips = slope * bills + intercept
```

```python
pred_tips
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,4))
plt.scatter(bills,tips)
plt.plot(sorted(bills),sorted(tips),'r--')
plt.plot(bills,pred_tips)
plt.xlabel('Bills')
plt.ylabel('Tips')
plt.show()
```
