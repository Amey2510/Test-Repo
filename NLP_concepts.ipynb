Text Tokenization,Stopword Removal,Stemming and Lammenization

```python
data = '''
Data analysis is the process of inspecting, cleansing, transforming, and modeling data with the goal of discovering useful information, informing conclusions, and supporting decision-making.[1] Data analysis has multiple facets and approaches, encompassing diverse techniques under a variety of names, and is used in different business, science, and social science domains.[2] In today's business world, data analysis plays a role in making decisions more scientific and helping businesses operate more effectively.[3]
'''
```

```python
data
```

Data Cleaning

---



```python
import re
data = data.lower()
pattern = r'[^a-z .]'
new_data = re.sub(pattern,'',data)
new_data
```

Tokenization
word_tokens and sentence_tokens

```python
from nltk.tokenize import sent_tokenize,word_tokenize
sent_tokens = sent_tokenize(new_data)
sent_tokens
```

```python
len(sent_tokens)
```

## Unigrams => Word tokens are generated for single words

```python
word_tokens = word_tokenize(new_data)
```

```python
word_tokens
```

```python
len(word_tokens)
```

## Bigrams, Trigrams and Ngrams

```python
bi_tokens = list(nltk.bigrams(word_tokens))
```

```python
bi_tokens
```

```python
n_tokens = list(nltk.ngrams(word_tokens,n=6))
n_tokens
```

## StopWord Removal

```python
word_tokens
```

```python
from nltk.probability import FreqDist
f1 = FreqDist()
```

```python
for word in word_tokens:
  f1[word]+=1
f1
```

```python
f1.most_common()[:10]
```

```python
## StopWord Removal
from nltk.corpus import stopwords
stopwords.fileids()
```

```python
stop = stopwords.words('english')
stop
```

```python
filtered_words=[]
for word in word_tokens:
  if word not in stop:
    filtered_words.append(word)
filtered_words
```

```python
len(filtered_words)
```

# Stemming

## 1) Porter Stemmer

```python
from nltk.stem import PorterStemmer
words = ["easily","walking","happiness","given","marked","inspecting"]
ps = PorterStemmer()
ps_words = [ps.stem(word) for word in words]
ps_words
```

# 2) Lancaster Stemmer

```python
from nltk.stem import LancasterStemmer
ls = LancasterStemmer()
ls_words = [ls.stem(word) for word in words]
ls_words
```

# 3) SnowBall Stemmer

```python
from nltk.stem import SnowballStemmer
ss = SnowballStemmer('english')

ss_words = [ss.stem(word) for word in words]
ss_words
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()

lm_words = [lemma.lemmatize(word,pos='v') for word in words]
lm_words
```

```python
from nltk import pos_tag
for word in words:
  print(pos_tag([word]))
```
