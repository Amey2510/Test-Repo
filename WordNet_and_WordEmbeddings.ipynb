```python
from warnings import filterwarnings
filterwarnings('ignore')
from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import gutenberg
```

```python
text = gutenberg.raw("shakespeare-hamlet.txt")
sentences = sent_tokenize(text)
hamlet = [word_tokenize(sent.lower()) for sent in sentences]
```

## Using CBOW Method Continuous BagofWords

```python
import gensim
from gensim.models import Word2Vec
model1 = gensim.models.Word2Vec(hamlet,min_count=1,vector_size=100,window=5,sg=0)
```

```python
model2 = gensim.models.Word2Vec(hamlet, min_count=1, vector_size=100,
                                window=5, sg=1)
```

```python
print("Cosine similarity between 'william' " +
      "and 'shakespeare' - CBOW : ",
      model1.wv.similarity('william', 'shakespeare'))

print("Cosine similarity between 'actus' " +
      "and 'primus' - Skip-Gram : ",
      model2.wv.similarity('actus', 'primus'))
```
